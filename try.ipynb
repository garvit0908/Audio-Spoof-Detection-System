```python
import librosa
import numpy as np
import librosa.display
import matplotlib.pyplot as plt

# Load audio file (Ensure sample rate is high, e.g., 96000 Hz)
audio_path = "0119.wav"
y, sr = librosa.load(audio_path, sr=None)  # Keep original sample rate

# Compute STFT
stft = np.abs(librosa.stft(y))

# Get frequency bins
freqs = librosa.fft_frequencies(sr=sr)

# Find indices for frequencies above 20 kHz
high_freq_indices = np.where(freqs > 20000)[0]

# Compute energy in the high-frequency range
high_freq_energy = np.sum(stft[high_freq_indices, :], axis=0)

# Plot energy over time
plt.figure(figsize=(10, 4))
plt.plot(high_freq_energy, label="Energy above 20 kHz")
plt.xlabel("Time Frames")
plt.ylabel("Energy")
plt.title("High-Frequency Energy (>20 kHz)")
plt.legend()
plt.show()

# Check if high-frequency content exists
if np.max(high_freq_energy) > 0:
    print("Audio contains frequencies above 20 kHz.")
else:
    print("No significant frequencies above 20 kHz detected.")
```
